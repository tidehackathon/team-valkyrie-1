```python
import pandas as pd

df = pd.read_csv('overall.csv')
df.head()
```

```python
df2 = pd.crosstab(df['consumer_id'], df['provider_id'])
df2.head()
```

```python
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res

basket_input = df2.applymap(encode)
basket_input
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


frequent_itemsets = fpgrowth(basket_input, min_support=0.6, use_colnames=True)
# frequent_itemsets = apriori(basket_input.astype(bool), min_support=0.5, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift")

rules.head()
```

```python
rules = rules[rules['antecedents'].apply(list).apply(len) == 1]

rules['antecedents'] = rules['antecedents'].apply(list).apply(lambda x: x[0] if len(x) == 1 
                                                              else '-'.join([str(arg) for arg in x])).astype(str)
```

```python
rules['consequents'] = rules['consequents'].apply(list).apply(lambda x: x[0] if len(x) == 1
                                                              else '-'.join([str(arg) for arg in x])).astype(str)
```

```python
result = rules.query('confidence > 0.85').groupby('antecedents')['consequents'].apply(lambda x: '-'.join(x)
                                                                                     ).reset_index()
result
```

```python
result.to_csv('output.csv', index=False)
```
